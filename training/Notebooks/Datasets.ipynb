```python
import pandas as pd
from tdc.single_pred import ADME
```

```python
from indigo import *
indigo = Indigo()
```

```python
def get_data(n):
  data = ADME(name = n)
  split = data.get_split()
  #print(split)
  df = pd.concat([pd.DataFrame.from_dict(split['train']), pd.DataFrame.from_dict(split['valid']), pd.DataFrame.from_dict(split['test'])],ignore_index=True)
  return df
```

```python
def get_canon_smile(molecule):
    try:
        canon_smile = indigo.loadMolecule(molecule)
        canon_smile.aromatize()
        return canon_smile.canonicalSmiles()
    except IndigoException as e:
        print(f"Error: {e}")
        return ""
```

# Ashmika Data

```python
aqsol = get_data('Solubility_AqSolDB')
aqsol["Drug_canon"] = aqsol["Drug"].apply(get_canon_smile)
ash=aqsol
ash
```
